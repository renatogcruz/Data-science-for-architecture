```python
from sklearn.datasets import fetch_20newsgroups
# Escolher duas categorias para trabalhar, diferentes das usadas neste exemplo (sci.med e sci.space)
# alt.atheism, comp.graphics, comp.os.ms-windows.misc, comp.sys.ibm.pc.hardware, comp.sys.mac.hardware, 
# comp.windows.x, misc.forsale, rec.autos, rec.motorcycles, rec.sport.baseball, rec.sport.hockey, sci.crypt, 
# sci.electronics, sci.med, sci.space, soc.religion.christian, talk.politics.guns, talk.politics.mideast, 
# talk.politics.misc, talk.religion.misc
categories = ['sci.med', 'sci.space']
# Fará o download dos dados - é preciso estar conectado na internet
twenty_sci_news = fetch_20newsgroups(categories=categories)
```

```python
# Preparação para criar as features ou colunas das nossas linhas. Cada linha é um documento diferente. 
# Cada coluna é relacionada a uma palavra diferente.
import os
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

# Conta quantas vezes cada palavra aparece no documento
count_vect = CountVectorizer()
# Pesa a frequência na qual a palavra aparece no documento vs quanto aparece em todos os outros documentos
tfidf_transformer = TfidfTransformer()

word_count = count_vect.fit_transform(twenty_sci_news.data)
word_tfidf = tfidf_transformer.fit_transform(word_count)
word_list = count_vect.get_feature_names()

for n in word_tfidf[0].indices:        
    print("Word:", word_list[n], "has tfidf", word_tfidf[0, n])
```

```python
import numpy as np
type(twenty_sci_news.target[0])
```

```python
# Cria os arquivos de treino e teste
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(word_tfidf,twenty_sci_news.target,test_size=.3)
```

```python
# Treina o classificador SVM
from sklearn import svm
s = svm.SVC(kernel='poly', C=1).fit(X_train,y_train)
s
```

```python
# Testa o classificador
ypred = s.predict(X_test)
```

```python
# Quantifica a qualidade do treino vs teste
from sklearn import metrics

accuracy = metrics.accuracy_score(y_test, ypred)
precision = metrics.precision_score(y_test, ypred)
recall = metrics.recall_score(y_test, ypred)
```

```python
print("Acurácia:", accuracy, " Precisão:", precision, " Recall:", recall)
```

.iloc() localiza variáveis pelos índices das linhas e colunas

.loc() localiza variáveis pelos índces das linhas e nomes da coluna

````python

#df['floors'] == 3.5 # retorna booleanos True ou False

#df[df['floors'] == 3.5] # retorna todas as linhas TRUE #df[condição]

df[df['floors'] == 3.5][['floors', 'id']] # retorna todas as linhas TRUE #df[condição][lista de colunas que interessam]
````

````python
# 3 - Criar uma classificação para separar os imóveis em baixo padrão (<540,000.00) e alto padrão (>540,000.00).

# criando nova coluna e atribuindo standard para todas as linas
df['level'] = 'low_standard'
````

````python
# testando se a criação acima deu certo

df.head()
````

````python
# retorna booleanos True e False de acordo com a condição

df['price'] > 540000
````

````python
# o comando df.loc[df['price'] > 540000, 'level'] seleciona todas as linhas True da coluna 'level'
# depois estas colunas recebem a string 'high_standard'

df.loc[df['price'] > 540000, 'level'] = 'high_standard'
````

````python
df.head()
````

````python
# 4 - Relatório ordenado pelo valor e contendo as seguintes informações: id, data disponível, número quartos, 
# tamanho terreno, valor, classificação.

report = df[['id', 'date', 'price', 'bedrooms', 'sqft_lot', 'level']]

report.head()
````

````python
# ordenando por valor
df_ordenado = report[['id', 'date', 'price', 'bedrooms', 'sqft_lot', 'level']].sort_values('price', ascending=False)

df_ordenado.head()
````

````python
df_ordenado.to_csv('dataset_ordenado.csv', index=False)
````

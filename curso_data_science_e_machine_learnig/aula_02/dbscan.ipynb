# DBSCAN Practice - Make moons

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn.neighbors import NearestNeighbors
```

### Make moons data

```python
data = datasets.make_moons(n_samples=2000, noise=0.1)
X = data[0]

plt.scatter(X[:,0], X[:,1], s=20, alpha=0.7)
```

### Kmeans

```python
model = KMeans(n_clusters = 2)
labels = model.fit_predict(X)
plt.scatter(X[:,0], X[:,1],c=labels, cmap='viridis', s=20, alpha=0.7)
```

### Optimal epsilon for DBSCAN - Elbow Method 

```python
neigh = NearestNeighbors(n_neighbors=5)
nbrs = neigh.fit(X)
distances, indices = nbrs.kneighbors(X)

distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

### DBSCAN

```python
dbscan = DBSCAN(eps=0.07, min_samples=5)
labels = dbscan.fit_predict(X)
plt.scatter(X[:,0], X[:,1],c=labels, cmap='viridis', s=20, alpha=0.7)
```
